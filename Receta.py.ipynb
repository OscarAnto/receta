# Importamos Archivo para pivote

```python
Data = pd.read_csv("C:/Users/user/Desktop/Mantenimiento Predictivo/Caso de uso 3_ Predictive Maintenance/maintenance_part_consumption.csv")
```

```python
Data.head()
```

```python
pivote = Data.pivot_table( index='Asset',columns = 'Reason', values='Quantity', aggfunc = 'sum', margins=True ).fillna('0')
```

```python
pivote.head(15)
```

```python
pivote.info()
```

```python
pivote = pivote.astype(int)
```

```python
pivote.info()
```

```python
#Se realizo el cambio de nombre añadiendo el sufijo _Quantity_sum a las columnas creadas
pivote.columns = [f'{i}_Quantity_sum' for i in pivote.columns]
#Se resetea el indice para que asset sea columna
pivote = pivote.reset_index()
#Se elimina la columna All que no es necesaria
del(pivote['All_Quantity_sum'])
```

```python
pivote.head(15)
```

# Tercer archivo creado

```python
pivote.to_csv (r'C:/Users/user/Desktop/maintenance_part_consumption_by_Reason.csv', index=None, header = True)
```

# Carga de los csv para hacer el join 

```python
maintenance_failures = pd.read_csv("C:/Users/user/Desktop/Mantenimiento Predictivo/Caso de uso 3_ Predictive Maintenance/maintenance_failures.csv")
```

```python
maintenance_failures.head()
```

```python
usage_data_by_Asset_prepared = pd.read_csv("C:/Users/user/Desktop/usage_data_by_Asset_prepared.csv")
```

```python
usage_data_by_Asset_prepared.head()
```

```python
maintenance_part_consumption_by_Reason = pd.read_csv("C:/Users/user/Desktop/maintenance_part_consumption_by_Reason.csv")
```

```python
maintenance_part_consumption_by_Reason.head()
```

```python
#Se realiza un left join agregando lo que hay en usage_data a maintance_failure
result = pd.merge(maintenance_failures, usage_data_by_Asset_prepared, on = "Asset")

#Al resultado del join anterior se le agrega maintenance_part
result = pd.merge(result, maintenance_part_consumption_by_Reason, on = "Asset")
```

```python
result.head()
```

```python
result.to_csv (r'C:/Users/user/Desktop/Asset_modelisation_table.csv', index=None, header = True)
```
